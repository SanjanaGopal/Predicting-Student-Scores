### 1) Results

Our model obtained can be represented by the equation:

###### Score = 1.98942311 + 9.83544438 (Hours of Study)

where intercept = 1.98942311 and coefficient = 9.83544438

This means that there is a positive relationship between the hours that a student studies and the scores they get.

From our initial exploration we observed the same trend. It was further observed that more students tended to study for moderate and long durations of time rather than short durations.


This model obtained a Mean Absolute Error (MAE) of 4.140342096254413 on the test data.
Despite the small nature of the dataset, the model seems reasonable. We also tried to input a random value for Hours to obtain a predicted Score. An input of 6 hours gave a score of 61.

### 2) Method

The data consists of 2 variables - Hours and Score. After getting a basic idea about the dimension and features of the dataset, we proceeded for an Exploratory Data Analysis (EDA) through a univariate and bivariate analysis. We used descriptive statistics and visualisation for this stage.

Univariate analysis with 'Hours' saw no outliers:

From 'Hours' we created a new variable called 'Duration' by bucketing the hours into 3 categories  - 1 to 3 hours, 3 to 6 hours and greater than 6 hours. This was just to add some granularity to our analysis. Our second univariate analysis was done with this variable.

For Bivariate analysis, we obtained a scatterplot showing the trend of 'Scores' with 'Hours' of study. Further we added 'Duration' as a hue to analyse with another level.

Data was then prepared for modeling by marking the dependent, independent variables and splitting data into train and test with a ratio of 80:20.

Used sklearn to obtain Linear Regression model on train data and plotted corresponding regression line. Then we used the model to predict on test data and used Mean Absolute Error (MAE) as the metric to evaluate the model.
